```python
import pandas as pd
import numpy as np

# Simulation of what Task 5 & 6 will produce
data = {
    'Month': ['Sept', 'Oct', 'Nov', 'Dec'],
    'Actual_Expense': [3200, 3100, 3050, 3500],
    'Prophet_Pred': [3150, 3180, 3100, 3400],
    'ANN_Pred': [3210, 3090, 3060, 3480],
    'Category': ['Food', 'Rent', 'Food', 'Shopping'],
    'Category_Amt': [800, 1500, 750, 1200]
}

df = pd.DataFrame(data)
```

```python
from sklearn.metrics import mean_absolute_error

# Calculate which model is better
mae_prophet = mean_absolute_error(df['Actual_Expense'], df['Prophet_Pred'])
mae_ann = mean_absolute_error(df['Actual_Expense'], df['ANN_Pred'])

print(f"Prophet Error: ${mae_prophet:.2f}")
print(f"ANN Error: ${mae_ann:.2f}")

# This logic picks the 'winner' for your report
best_model = "ANN" if mae_ann < mae_prophet else "Prophet"
print(f"The model you should trust for advice is: {best_model}")
```

```python
# Calculate trend for the LLM
last_actual = df['Actual_Expense'].iloc[-1]
predicted_next = df['ANN_Pred'].iloc[-1]
pct_change = ((predicted_next - last_actual) / last_actual) * 100
top_cat = df['Category'].iloc[-1]

# This is your 'Advisory Prompt'
prompt = f"""
As a financial advisor, look at this:
- Spending is predicted to change by {pct_change:.1f}%.
- The biggest spending category is {top_cat}.
Provide 3 tips to help the user stay on budget.
"""
print(prompt)
```
